```python
import pandas as pd
import seaborn as sns
import pylab as plt
%matplotlib inline
```

```python
df = pd.read_csv("./data/Error by type of visualization (wwd2018) (Responses) - Form Responses 1.csv")
df.head()
```

```python
variables_already_presents = ["Timestamp"]
columns_combine = ["Saturation","Lightness","Areas (circles)","Areas (rectangles)","Length (bars)"]
df_tidy = pd.melt(df,
       id_vars=variables_already_presents,
       value_vars=columns_combine,
       var_name="Type",
       value_name="Number")
df_tidy
```

```python
import numpy as np
df_tidy["Error"] = df_tidy["Number"]/8
df_tidy.head()
```

```python
sns.violinplot(x="Type",y="Error",data=df_tidy,color="gray",inner="quartiles")
plt.xticks(rotation=70)
plt.yscale("log")
plt.ylim(0.1,5)
```
